```python
import pandas as pd
```

```python
def intersection(lst1, lst2):
    lst3 = [value for value in lst1 if value in lst2]
    return lst3
```

```python
domains_mapping = pd.read_csv("domains-features-mapping.csv");
print(domains_mapping)
```

```python
tasks_mapping = pd.read_csv("tasks-features-mapping.csv");
print(tasks_mapping)
```

```python
detected_domain = "computer vision"
domain_features = []

for index in domains_mapping.index:
  if domains_mapping['Domain'][index].lower() == detected_domain.lower():
    domain_features.append(domains_mapping['Feature'][index])
print(domain_features)
```

```python
detected_tasks = ["clustering", "classification", "task assignment"]
out_features = {}
tmp = []

for task in detected_tasks:
  tmp = []

  for index in tasks_mapping.index:
    if tasks_mapping['Task'][index].lower() == task.lower():
      tmp.append(tasks_mapping['Feature'][index])
    
  print(tmp)
  out_features[task] = intersection(tmp, domain_features)


```

```python
for task in detected_tasks:
  print("Domain: {} - Task: {} - Sensitive Features: {}".format(detected_domain, task, out_features[task]))
```
