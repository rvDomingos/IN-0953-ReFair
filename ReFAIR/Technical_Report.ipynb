# *ReFAIR: Toward a Context-Aware Recommender for Fairness Requirements Engineering*

Paper Abstract: Machine learning (ML) is increasingly being used as a key component of most software systems, yet serious concerns have been raised about the fairness of ML predictions. Such an increasing use of machine learning may have implications on software fairness: the inappropriate treatment of sensitive features may indeed lead machine learning algorithms to act unfairly with respect to minorities. Researchers have been proposing novel methods to support the development of fair machine learning solutions.
    
Nonetheless, most of them can only be used in late development stages, e.g., during model training, while there is a lack of methods that may provide practitioners with early fairness analytics enabling the treatment of fairness throughout the development lifecycle. This paper proposes **ReFair**, a novel context-aware requirements engineering framework that allows to classify sensitive features from User Stories. By exploiting natural language processing and word embedding techniques, our framework first identifies both the use case domain and the machine learning task to be performed in the system being developed; afterward, it recommends which are the context-specific sensitive features to be considered during the implementation. We assess the capabilities of **ReFair** by experimenting it against a synthetic dataset---which we built as part of our research---composed of 12,401 User Stories related to 34 application domains. Our findings showcase the high accuracy of **ReFair**, other than highlighting its current limitations.

# **(1) On the Validity of the Synthetic Dataset.**

In order to build and validate a dataset of synthetic user stories as realistic and complete as possible, we followed different methodological steps, which we detail below.

First, for the creation of user stories as closely akin to a precise application domain and machine learning technique as possible, we systematically queried chat-gpt for each domain and machine learning technique pair of the augmented taxonomy we created. The creation of synthetic user stories was subject to an incremental and continuous review process, partucularly every time a new US was generated, the first author of the paper manually verified its consistency and degree of realism, being therefore able to discard the USs whose quality was low.

 As a result of this manual validation, all the USs connected to 81 of the machine learning tasks originally considered were discarded, as they were deemed too specific for generating user stories having a structure and an expressiveness close to what a real practioner would have produced. Upon completion of the generation process, the second author of the article double-checked the operations conducted by the first author on a statistically significant sample of the synthetic USs (confidence level=95\%, margin of error=5\%).  Finally we then successfully created *12401* user stories connected to 34 application domains and 367 low-level or high-level machine learning techniques.

Considering that the synthetic user story dataset was created through the use of an LLM, each individual response received by the bot was evaluated on a case-by-case basis, chiding to reiterate in the generation in case of excessive similarity or overly simplistic responses. ##Dataset validation
Subsequently, our primary goal was to validate the synthetic dataset in order to ensure how closely the generated user stories could be aligned with those formulated by industry experts. Since validating an entire dataset of 12,401 user stories through qualitative techniques was not easily feasible, we decided to randomly select a statistically significant sample that could best represent the entire dataset as a whole.

In order to determine the minimum number of user stories required for empirical validation, we applied the following formula, where N = population size • e = margin of error (expressed as a percentage in decimal form) • z = z-score.

https://www.surveymonkey.com/mp/sample-size-calculator/





image.png

We considered a confidence level of the 95%, i.e., a z-score = 1.96 and a margin of error of the 5%. Obtaining 373 user stories as a minumum number of user story to valiate, in order to obtain a statistical significance for the overall user story dataset.

 In light of this, we decided to involve developers with proven experience in machine learning by using an online survey, disseminating a survey that asked them to evaluate 375 user stories in terms of:
1.   **Comprehensibility**: How easily understandable is the intrinsic text of the user story from a semantic-lexical point of view?

2.   **Actionability**: How useful is the informational content of the user story for the implementation of a related ML-intensive system?
3. **Realism**: How similar is the synthetic user story compared to how it would be formalized in a professional environment?

For each of these 3 indicators, we asked practitioners to express their opinions in terms of a 5-points likert-scale value, where each value corresponds to: 1: not at all; 2: slightly; 3: neutral; 4: very; 5: extremely.



In addition, for each user story we formalized an open-ended question, to ask practitioners how they should improve the unclear and unrealistic stories, providing us with practical suggestions or alternative USs.

We divided the 375 user stories to evaluate 75 different online questionnaire (5 user stories for questionnaire), that we further administered by using the Prolific dissemination platform. We asked two different practitioners to evaluate the same group of user stories, allowing for cases of overlap.

#**(2): Machine Learning Classifiers: Details on Evaluation Metrics and Results**



In this paragraph, we added further details regarding the experimentation and the analysis of main classification tasks enclosed in the refair pipeline, i.e., the Application domain and ML Tasks classifiers. In particular, we reported details regarding, the specific classification problem, the main results for both the classification problems.

## **Application Domain Classifier**


***Problem***: We defined the USs domain classification problem as a multiclass classification problem, i.e., a classification problem in which the main goal is to individuate a *single classification* label classifying among three or more classes (classifying instances into one of two classes is called binary classification), a simple example consist in classify a set of images of fruits which may be oranges, apples, or pears.

***Experiments and Metrics***: To evaluate the performances of this kind of classifier, we used the  formal [10 fold cross validation](https://machinelearningmastery.com/k-fold-cross-validation/) method, evaluatind 125 different combinations of Multi-class Classifiers and Embedding techniques (as discussed in the research peper), finally we selected the better one combination evaluating the following metrics widly used in the context of Multi-class classification problem. Partucularly we identified as the main evaluation metrics the following:


1.   *Accuracy*: in the constext of Multi-Class classification problem measures the proportion of correctly classified cases from the total number of instances in the dataset. This metrics is computed by dividing the number of correct predictions by the total number of the model predictions;
2.   *Precision*: in multi class classification problem consists in the average of the precisions of each output class of the problem. For each single output class X, the precision is defined as the fraction of instances of **X** correctly classified as belonging to a specific class out  (True Positive - TP<sub>Class X</sub>) of all instances the model predicted to belong to that class **X** (True Positive + False Positive - TP<sub>Class X</sub> + FP<sub>Class X</sub>);
3. *Recall*: in multi-class classification the recall of a certain class X is the fraction of instances in a class that the model correctly classified  (True Positive - TP<sub>Class X</sub>) out of all instances in that class, that are correctly or wrongly classified (True Positive + False Negative - TP<sub>Class X</sub> + FN<sub>Class X</sub>). As the precision, the general recall of the classifier of a Multi-Class model consists in the average of the recalls of the all of each output class of the ML problem.
4. *F1-Score*: that consists in the harmonic mean of the Multi-class classifier's precision and recall, which provides a balance between these two measures:






image.png

Further details regarding those evaluation metrics are available are well described in this external resource: [Accuracy, precision, and recall in multi-class classification](https://www.evidentlyai.com/classification-metrics/multi-class-metrics).

**Experimental Results**: To select the better one combination of ML model and embedding techniques for the ReFair domain detection task, we performed the 10-fold cross validation for each of the 125 combination of the classifiers and embeddings techniques. In details, we performed the model training on each of the 10 folds of our USs dataset, and we mesured the evaluation metrics for each fold.

Here an example of the 10-Folds evaluation reports that we released in our online appendix:



image.png

Finally, we aggregate the single folds results, computing the average Accuracy and F1-Score for each combination of ML task and Embedding technique. Selecting as ReFair's application domain classifier the combination of embedding technique and algorithm that obtained the best evaluation results.
Here the average results for the best 15 combinations of ML Tasks and embedding techniques.

## **Machine Learning task Classifier**


***Problem***: We defined the USs machine learning classification problem as a multilabel classification problem, i.e., a classification problem in which the main goal is to predict zero or more class labels for each input. Unlike normal classification tasks where class labels are mutually exclusive, multi-label classification requires specialized machine learning algorithms that support predicting multiple mutually non-exclusive classes or “labels.”

***Experiments and Metrics***: To evaluate the performances of this kind of classifier, we used the  formal [10 fold cross validation](https://machinelearningmastery.com/k-fold-cross-validation/) method, evaluatind 90 different combinations of Multi-class Classifiers and Embedding techniques (as discussed in the research peper), finally we selected the better one combination evaluating the following metrics widly used in the context of Multi-Label classification problem. Partucularly we identified as the main evaluation metrics the following:


1.   *Precision*: In multi-label classification problem, the precision of a ML Model is It is the propotion of predicted correct labels to the total number of predicted labels, averaged over all instances.

image.png


2. *Recall*: in multi-label classification problems the recall is defined as the proportion of predicted correct labels to the total number of actual labels, averaged over all instances.


image.png

4. *F1-Score*: As the a multi-class classification problem, the F1-Score consists in in the harmonic mean of the Multi-Label classifier's precision and recall, which provides a balance between these two measures:

image.png

4. *Hamming Loss*: This metric reports how many times on average a multi-label classifier wrongly assign a label to a certain individual of the validation set.Therefore, hamming loss takes into account the prediction error (an incorrect label is predicted) and missing error (a relevant label not predicted), normalized over total number of classes and total number of examples.

image.png


Further details regarding those evaluation metrics are available are well described in this external resource: [Metrics for Multilabel Classification](https://mmuratarat.github.io/2020-01-25/multilabel_classification_metrics).

**Experimental Results**: As the domain classification problem, To select the better one combination of ML model and embedding techniques for the ML tasks classification task, we performed the 10-fold cross validation for each of the 90 combination of multi-label classifiers and embeddings techniques. In details, we performed the model training on each of the 10 folds of our USs dataset, and we mesured the evaluation metrics for each fold.

Here an example of the 10-Folds evaluation reports that we released in our paper appendix:

image.png

Finally, we aggregate the single folds results, computing the average F1-Score and Hamming Loss for each combination of ML task and Embedding technique. Selecting as ReFair's application ml classifier the combination of embedding technique and algorithm that obtained the best evaluation results. Here the average results for the best 15 combinations of ML Tasks and embedding techniques.

image.png

As discussed in the main Research paper, we finally selected the combination of Label Powerset, Glove, and Linear Support Vector Classification established
an F1-Score slightly above 90%, with a Hamming Loss of 5% on the ML tasks classification task.

# **(3) ReFair: Further Insights on Its Validation.**

Finally, we have included additional details about the overall accuracy of ReFair's predictions. We considered how well the entire machine learning pipeline produces correct predictions in terms of sensitive features for each user story in the synthetic dataset. Please considered that the main results concerning the single tasks of domain and task classification were discusses in the experimental section of our research paper.

To properly evaluate the overall ReFair predictions, we started with considering our prototype as a complete and unique black box model, and in reason of that we design a first validation method in order to evaluate what is the set of features our ReFair pipeline suggests independently from the predicted domain and machine learning tasks.

Firstly, we generated an ideal oracle, identifing the for each user story, the ideal set of features to suggest in relation to the application domain and machine learning tasks that we used to generate them. In details, for each user story of the dataset, we queried the domains -> features and ML tasks -> features mappings, that we formalized before dataset generation (Fabris et al. Alghoritmic fairness dataset: the story so far 2022), in order to obtain the ideal set of unique features that our ReFair approach **should** suggest.

Here an example of User Story oracle prediction:


1.   **User Story**: As a computer vision researcher, I want to use learning vector quantization to classify objects in images based on their features;  
2. **Sensitive Features**: {'geography', 'gender', 'race', 'age', 'skin tone'}.

Successively, we ran ReFair on each User Story of our synthetic dataset to save the **actual** sensitive features that our approach suggests.

Here te example of actual feature predicted by ReFair on the same User Story:


1.   **User Story**: As a computer vision researcher, I want to use learning vector quantization to classify objects in images based on their features;
2. **Sensitive Features**: {'race', 'skin tone', 'age', 'gender', 'geography'}.

For both, oracle and actual ReFair prediction we stored the sensitive features related to each user story in a structured spreadsheet, available in our paper appendix.

We then ran ReFair on each US in our dataset to automatically evaluate the cases in which ReFair produced correct, partially correct,and incorrect sets of features with respect to the oracle by leveraging the  [token-level Levenshtein distance](https://medium.com/@ethannam/understanding-the-levenshtein-distance-equation-for-beginners-c4285a5604f0), considering a token representing a single feature. To compute the distance, we leveraged the nltk implementation.

Here the detailed Results we obtained :



1.   Empty set of feature in the oracle: 7457 - 60% of the overall suggestions;
2.   Empty set of feature in the Refair output: 7416 - 60% of the overall suggestions;
3.   (Oracle, Refair output) pairs providing the same set of features: 11969 - 97% of the overall suggestions;
4. (Oracle, Refair output) pairs differing in the set of features proposed in 1 feature: 41 - 0.3% of the overall suggestions;
5.  (Oracle, Refair output) pairs differing in the set of features proposed in 2 features: 70 - 0.6% of the overall suggestions;
6. (Oracle, Refair output) pairs differing in the set of features proposed in more than 2 features: 321 - 2.6% of the overall suggestions;
7. Empty set for both: 7253 - 65% of the overall suggestions;
8. Instances in which the oracle set of feature is not empty: 4944 - 39% of the overall suggestions.
  1. (Oracle, Refair output) pairs providing the same set of features with at least 1 feature: 4716 - 95% of the not empty suggestions;
  2. (Oracle, Refair output) pairs providing the same set of features with at least 1 feature differing in 1 feature: 26 - 0.5% of the not empty suggestions;
  3. (Oracle, Refair output) pairs providing the same set of features with at least 1 feature differing in 2 features: 44 - 0.9% of the not empty suggestions;
  4. (Oracle, Refair output) pairs providing the same set of features with at least 1 feature differing in more than 2 features: 158 - 3.2% of the not empty suggestions.




Sequentially, appling an additional evaluation based on set difference operations, in order to understand the main causes of errors among the 432 wrong predictions highlited by the Levenshtein distance:


1.   Feature suggestions totally wrong due void oracle: 204 - 47%
2.   Features suggestions totally void rather than not void oracle 217 - 50%
3.   Feature suggestion totally different from a not void oracle: 11 - 2.5%

To assess the ReFair validation results in terms of differences between ReFair prediction and the Oracle, we evaluated the Jaccard and the Mean sets overlap indexes, obtaining the following mean results:



1.   Mean overlap - 0.9610753972632007;
2.   Jaccard overlap - 0.9631388652474817;
3.   **Mojo distance** - 0.037892868744658796.





------------------------------------------------------------------------------------

Despite the general validation of the entire set of feature suggested, it is worth remarking that the predictions of our ReFair pipeline are completely correct only when the application domain and learning tasks perfectly corresponded with the predictions proposed by the oracle, which we formalized in our paper's appendix.

To further investigate this aspect, we proposed an alternative analysis that take into account the inner component predictions of our ReFair pipelines, i.e., the application domain and the machine learning tasks related to a specific user story.


In details, we constructed an alternative validation oracle, that report for each user story, the following 3 elements:

1.   The correct application domain of the user story;
2.   The ideal machine learning tasks to develope in practice the user story (i.e., the ideal tasks related to the specific ML techniques we used to generate the story);
3.   The sensitive features that our ReFair pipeline should suggest in relation toboth the learning task and the application domain of the story

Sequentially we stored the same information with respect to the actual ReFair's predictions, and than we evaluated in how many cases the ReFair suggestions are consistent with the oracle, as further explained.

Here an example of User Story oracle:


1.   **User Story**: A healthcare provider wants to develop a bidirectional recurrent neural network that can predict patient outcomes based on medical history, lifestyle factors, and environmental conditions;
2. **Application domain**: Health;
3. **Task & Features**: {'classification': ['age', 'geography', 'sex', 'ethnicity', 'gender'], 'regression': ['age', 'ethnicity', 'gender', 'geography', 'sex'], 'matching': ['age', 'gender', 'geography'], 'representation learning': ['age', 'ethnicity', 'gender', 'geography', 'sex']}



Finally we manually evaluated the similarity and the differences among ReFair and the Oracle.
Below are the detailed statistics regarding the predictions of ReFair compared to the oracle we constructed:





1. For **9991 user stories (81%)**, ReFair correctly suggests sensitive features by accurately predicting tasks and the application domain.
  *   3330 user story for witch the ReFair Predictions corresponds exactly with the oracle's;
  *   6661 cases in which ReFair correctly makes void predictions.

2.   For the **992 user stories (8%)** ReFair provides wrong suggestions.
  *  Cases in which ReFair wrongly predicts the application domain: 57;
  *  Cases in which ReFair did not predict any ML task wrongly: 213;
  *  Cases in which ReFair wrongly predicts completely the ml tasks: 722.
3. For **1418 user stories (11%)** ReFair suggests features that are partially correct, because the application domain is correctly predicted, but the tasks classifier suggested both application correct and uncorrect machine learning tasks.









In the end, we coutend the misclassification occurrencies for both, domain and ml tasks classification, observing that the most misclassified application domain and machine leanrning task are: (1) the "health" domain, and the (2) "classification" task.
